```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path='https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillUp/labs/ML-FinalAssignment/Weather_Data.csv'
```

```python
df = pd.read_csv(path)
df.head()
```

```python
df_sydney_processed = pd.get_dummies(data=df, columns=['RainToday', 'WindGustDir', 'WindDir9am', 'WindDir3pm'])
df_sydney_processed.head()
```

```python
df_sydney_processed.replace(['No', 'Yes'], [0,1], inplace=True)
df_sydney_processed.head()
```

```python
df_sydney_processed.drop('Date',axis=1,inplace=True)
```

```python
df_sydney_processed = df_sydney_processed.astype(float)
```

```python
features = df_sydney_processed.drop(columns='RainTomorrow', axis=1)
Y = df_sydney_processed['RainTomorrow']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,  y_test = train_test_split(features, Y, test_size = 0.2, random_state = 10)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
LinearReg = LinearRegression()
LinearReg
```

```python
LinearReg.fit(X_train, y_train)
```

```python
predictions = LinearReg.predict(X_test)
predictions[0:5]
```

```python
import sklearn.metrics as metrics
```

```python
LinearRegression_MAE = metrics.mean_absolute_error(y_test, predictions)
LinearRegression_MSE = metrics.mean_squared_error(y_test, predictions)
LinearRegression_R2 = metrics.r2_score(y_test, predictions)
print("LinearRegression_MAE: ", LinearRegression_MAE)
print("LinearRegression_MSE: ", LinearRegression_MSE)
print("LinearRegression_R2: ", LinearRegression_R2)
```

```python
Report = {"Metrics":["MAE","MSE","R2"],"Result": 
[LinearRegression_MAE,LinearRegression_MSE,LinearRegression_R2]}
pd.DataFrame(Report)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN = KNeighborsClassifier(n_neighbors=4)
KNN.fit(X_train, y_train)
KNN
```

```python
predictions = KNN.predict(X_test)
predictions[0:5]
```

```python
from sklearn.metrics import jaccard_score
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
KNN_Accuracy_Score = accuracy_score(predictions, y_test)
KNN_JaccardIndex = jaccard_score(predictions, y_test)
KNN_F1_Score = f1_score(predictions, y_test)
```

```python
Report = {"Metrics":["KNN_Accuracy_Score","KNN_JaccardIndex","KNN_F1_Score"],"Result": 
[KNN_Accuracy_Score,KNN_JaccardIndex,KNN_F1_Score]}
pd.DataFrame(Report)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
Tree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
Tree.fit(X_train, y_train)
```

```python
predictions = Tree.predict(X_test)
predictions[0:5]
```

```python
Tree_Accuracy_Score =  accuracy_score(y_test, predictions)
Tree_JaccardIndex =  jaccard_score(y_test, predictions)
Tree_F1_Score =  f1_score(y_test, predictions)

Report = {"Metrics":["Tree_Accuracy_Score","Tree_JaccardIndex","Tree_F1_Score"],"Result": 
[Tree_Accuracy_Score,Tree_JaccardIndex,Tree_F1_Score]}
pd.DataFrame(Report)
```

```python
X_train, X_test, y_train,  y_test = train_test_split(features, Y, test_size = 0.2, random_state = 1)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression(solver='liblinear')
LR.fit(X_train, y_train)
```

```python
predictions = LR.predict(X_test)
predictions[0:5]
```

```python
predict_proba = LR.predict_proba(X_test)
predict_proba[0:5]
```

```python
from sklearn.metrics import log_loss
LR_Accuracy_Score =   accuracy_score(y_test, predictions)
LR_JaccardIndex =  jaccard_score(y_test, predictions)
LR_F1_Score =  f1_score(y_test, predictions)
LR_Log_Loss = log_loss(y_test, predict_proba)
```

```python
Report = {"Metrics":["LR_Accuracy_Score","LR_JaccardIndex","LR_F1_Score", "LR_Log_Loss"],"Result": 
[LR_Accuracy_Score,LR_JaccardIndex,LR_F1_Score, LR_Log_Loss]}
pd.DataFrame(Report)
```

```python
Report = {"Metrics":["SVM_Accuracy_Score","SVM_JaccardIndex","SVM_F1_Score"],"Result": 
[SVM_Accuracy_Score,SVM_JaccardIndex,SVM_F1_Score]}
pd.DataFrame(Report)
```
